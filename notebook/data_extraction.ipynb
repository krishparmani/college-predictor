```python
import re

with pdfplumber.open(pdf_path) as pdf:
    text = pdf.pages[5].extract_text()

# College Code + College Name
college_match = re.search(
    r'^\s*(\d{4})\s*-\s*(.+)$',
    text,
    re.MULTILINE
)

# Branch Code + Branch Name
branch_match = re.search(r'(\d{9})\s*-\s*(.+)', text)

# Status
status_match = re.search(r'Status:\s*(.+)', text)

college_code = college_match.group(1) if college_match else None
college_name = college_match.group(2).strip() if college_match else None

branch_code = branch_match.group(1) if branch_match else None
branch_name = branch_match.group(2).strip() if branch_match else None

status = status_match.group(1).strip() if status_match else None

print("College Code :", college_code)
print("College Name :", college_name)
print("Branch Code  :", branch_code)
print("Branch Name  :", branch_name)
print("Status       :", status)
```

```python
rows = []

# Metadata
metadata = {
    "academic_year": "2023-2024",
    "college_code": college_code,
    "college_name": college_name,
    "branch_code": branch_code,
    "branch_name": branch_name,
    "status": status
}

# First table only
headers = table[0][1:]      # GOPENH, GSCH, GVJH...
values = table[1][1:]       # 74562\n(65.96)...

stage = table[1][0]         # I

for category, value in zip(headers, values):

    merit_rank = None
    percentile = None

    if value:
        parts = value.split("\n")

        if len(parts) == 2:
            merit_rank = parts[0]
            percentile = parts[1].replace("(", "").replace(")", "")

    rows.append({
        **metadata,
        "seat_category": category,
        "stage": stage,
        "merit_rank": merit_rank,
        "percentile": percentile
    })

df = pd.DataFrame(rows)

df
```

```python
def extract_metadata(text, academic_year):
    import re

    college_match = re.search(
        r'^\s*(\d{4})\s*-\s*(.+)$',
        text,
        re.MULTILINE
    )

    branch_match = re.search(
        r'(\d{9})\s*-\s*(.+)',
        text
    )

    status_match = re.search(
        r'Status:\s*(.+)',
        text
    )

    metadata = {
        "academic_year": academic_year,
        "college_code": college_match.group(1) if college_match else None,
        "college_name": college_match.group(2).strip() if college_match else None,
        "branch_code": branch_match.group(1) if branch_match else None,
        "branch_name": branch_match.group(2).strip() if branch_match else None,
        "status": status_match.group(1).strip() if status_match else None
    }

    return metadata
```
