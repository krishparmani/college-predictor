```python
import pandas as pd
```

```python
df23 = pd.read_csv("../dataset/processed/2023-2024.csv")
df24 = pd.read_csv("../dataset/processed/2024-2025.csv")
df25 = pd.read_csv("../dataset/processed/2025-2026.csv")
```

```python
print(df23.shape)
print(df24.shape)
print(df25.shape)
```

```python
print(df23.columns)
print(df24.columns)
print(df25.columns)
```

```python
merged_df = pd.concat(
    [df23, df24, df25],
    ignore_index=True
)

print(merged_df.shape)
```

```python
merged_df["academic_year"].value_counts()
```

```python
print("Duplicates:", merged_df.duplicated().sum())
```

```python
merged_df.to_csv(
    "../dataset/interim/merged_dataset.csv",
    index=False
)

print("Merged dataset saved successfully!")
```
